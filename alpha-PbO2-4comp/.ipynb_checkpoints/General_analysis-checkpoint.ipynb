```python
import os
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

kB = 8.6173303E-5 #eV/K

```

```python
file = "./summary_alphaPbO2_deltaHvsigma_ordered_Ei.csv"
df_alpha = pd.read_csv(file)
df_alpha
```

```python
def plot_columns(df, col_x, col_y, filter_col, filter_list,fname):
    """
    Plots col_x against col_y for rows where any of the strings in filter_list appear in filter_col.

    Parameters:
    - df: pandas DataFrame
    - col_x: str, column name to be plotted on the x-axis
    - col_y: str, column name to be plotted on the y-axis
    - filter_col: str, column name to apply the filter
    - filter_list: list of strings to filter filter_col

    Returns:
    - A scatter plot of col_x vs col_y for the filtered rows.
    """
    # Filter rows where filter_col contains any of the strings in filter_list
    filtered_df = df[~df[filter_col].astype(str).str.contains('|'.join(filter_list), na=False)]
    
    # Plot col_x against col_y
    plt.figure(figsize=(8, 6))
    plt.scatter(filtered_df[col_x], filtered_df[col_y], color='b', alpha=0.7, label=('Exclude '+','.join(filter_list)))
    plt.xlabel(col_x)
    plt.ylabel(col_y)
    plt.title(f"Plot of {col_x} vs {col_y} for Filtered Rows")
    plt.legend()
    plt.grid(True)
    plt.savefig(fname)
    plt.show()
```

```python
excl_list = ['Nb','Mn']
```

```python
plot_columns(df=df_alpha, col_x='Cation sigma', col_y= 'sigma bond length', filter_col='Formula', filter_list=excl_list,fname="scatter_Ei_v_bond_aPbO2.png")
```

```python
filtered_df = df_alpha[~df_alpha['Formula'].astype(str).str.contains('|'.join(excl_list), na=False)]
csvfilestr = "./summary_alphaPbO2_ordered.csv"
filtered_df.to_csv(csvfilestr, index=False)
filtered_df
```

```python
ncat=4

def H_to_T(eV):
    return (eV/(kB * np.log(ncat)))

def T_to_H(K):
    return (K * (kB * np.log(ncat)))

fig, ax = plt.subplots(figsize=(8, 8))

# Customizing the plot
ax.set_xlabel('Bond $\sigma$ ($\AA$)')
ax.set_ylabel('E$_i$ $\sigma$ (eV)')

ax.scatter(filtered_df['sigma bond length'], filtered_df['Cation sigma'], color='blue', s=100, alpha=0.7)

# ax.fill_betweenx([0.0, 0.1], 0.0, T_to_H(1800), color='green', alpha=0.3)

ax.legend()
ax.grid(True)

ax.set_xlim(0.0,0.20)
ax.set_ylim(0.0,0.12)

# # Top x-axis: temperature in K
# secax = ax.secondary_xaxis('top', functions=(H_to_T, T_to_H))
# secax.set_xlabel("Temperature (K)")

plt.rcParams.update({'font.size': 14})  # Set default font size

plt.savefig('./scatter_bond_v_Ei_aPbO2.png')

plt.show()
```
